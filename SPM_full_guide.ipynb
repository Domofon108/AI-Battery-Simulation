## 1. Introduction
This notebook contains a complete walkthrough of the Single Particle Model (SPM) using PyBaMM. It includes theory, simulation, visualization, and parameter inspection.

## 2. Import Libraries

```python
import pybamm
import matplotlib.pyplot as plt
```

## 3. Create the SPM Model

```python
model = pybamm.lithium_ion.SPM()
model
```

## 4. Load Parameter Set (Chen2020)

```python
params = pybamm.ParameterValues("Chen2020")
params
```

## 5. Build and Solve Simulation

```python
sim = pybamm.Simulation(model, parameter_values=params)
solution = sim.solve([0, 3600])
solution
```

## 6. Plot Key Outputs

```python
sim.plot()
```
