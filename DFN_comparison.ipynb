# DFN model.Setups comparison 

```python
import pandas as pd
import matplotlib.pyplot as plt
import pybamm
import plotly.express as px
import plotly.graph_objects as go 
from plotly.subplots import make_subplots
import numpy as np
```

## Single models

#### Chen2020 — NMC811 / Graphite

```python
parameters_values = pybamm.ParameterValues("Chen2020")
model = pybamm.lithium_ion.DFN()
```

```python
experiment = pybamm.Experiment(
    [
        (
            "Discharge at 1C until 3V",
            "Rest for 1 hour",
            "Charge at 1C until 4.2V",
            "Hold at 4.2V until C/50",
        )
    ]
)
sim = pybamm.Simulation(model, 
                        experiment=experiment, 
                        parameter_values=parameters_values)
sol = sim.solve()
```

```python
sol.plot()
```

```python
vars_to_export = ["Time [s]", "Terminal voltage [V]", "Current [A]"]
data = sol.get_data_dict(vars_to_export)

df = pd.DataFrame(data)
print(df.head())
```

```python
df.to_csv("Chen2020.csv", index=False)
```

## Parametries comparison

```python
presets = {
    "Chen2020":       pybamm.ParameterValues("Chen2020"),
    "OKane2022":      pybamm.ParameterValues("OKane2022"),
    "Chayambuka2022": pybamm.ParameterValues("Chayambuka2022"),
    "ORegan2022":     pybamm.ParameterValues("ORegan2022"),
    "Prada2013":      pybamm.ParameterValues("Prada2013"),
}
```

```python
rows = {}
for preset_name, params in presets.items():
    for key, value in params.items():
        if isinstance(value, (int, float)):          # only numbers, no functions
            if key not in rows:
                rows[key] = {}
            rows[key][preset_name] = value

df_params = pd.DataFrame(rows).T
df_params.index.name = "Parameter"

df_params.to_csv("DFN_parameters_comparison.csv")
df_params
```

```python
df_diff = df_params[df_params.nunique(axis=1) > 1]
print(f"Different parameters: {len(df_diff)}")
df_diff
```

```python
df_norm = df_diff.copy()
df_norm = df_norm.apply(lambda row: (row - row.min()) / (row.max() - row.min() + 1e-30), axis=1)

fig = px.imshow(
    df_norm,
    labels=dict(x="Preset", y="Parameter", color="Norm. value"),
    title="Parameter comparison across PyBaMM presets (normalized)",
    aspect="auto",
    color_continuous_scale="RdYlGn",
    template="plotly_dark",
)
fig.update_layout(height=900)
fig.show()
```

```python
baseline = "Chen2020"
df_pct = df_diff.copy()

for col in df_pct.columns:
    if col != baseline:
        df_pct[col] = ((df_pct[col] - df_pct[baseline]) / df_pct[baseline].abs() * 100).round(1)

df_pct[baseline] = df_diff[baseline]   # оставляем исходные значения базы
df_pct.columns = [f"{c} (% vs Chen2020)" if c != baseline else f"{baseline} (base)" 
                  for c in df_pct.columns]
df_pct
```

## Model comparison 

```python
dataframes = {
    "Chayambuka": pd.read_csv("Chayambuka2022.csv"),
    "Chen":       pd.read_csv("Chen2020.csv"),
    "OKane":      pd.read_csv("OKane2022.csv"),
    "Prada":      pd.read_csv("Prada2013.csv"),
    "ORegan":     pd.read_csv("ORegan2022.csv"),
}
```

```python
colors = ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]a
```

###  1. Voltage vs Time

```python
fig1 = go.Figure()
for (name, df), color in zip(dataframes.items(), colors):
    fig1.add_trace(go.Scatter(
        x=df["Time [s]"], y=df["Terminal voltage [V]"],
        mode="lines", name=name, line=dict(color=color, width=1)
    ))
fig1.update_layout(
    title="Terminal Voltage vs Time",
    xaxis_title="Time [s]", yaxis_title="Terminal Voltage [V]",
    template="plotly_dark", hovermode="x unified"
)
fig1.show()
```

### 2. Current vs Time

```python
fig2 = go.Figure()
for (name, df), color in zip(dataframes.items(), colors):
    fig2.add_trace(go.Scatter(
        x=df["Time [s]"], y=df["Current [A]"],
        mode="lines", name=name, line=dict(color=color, width=1)
    ))
fig2.update_layout(
    title="Current vs Time",
    xaxis_title="Time [s]", yaxis_title="Current [A]",
    template="plotly_dark", hovermode="x unified"
)
fig2.show()
```

### 3.  Number of points by cycles (coverage comparison)

```python
fig4 = go.Figure()
for (name, df), color in zip(dataframes.items(), colors):
    cycle_counts = df.groupby("Cycle").size().reset_index(name="count")
    fig4.add_trace(go.Bar(
        x=cycle_counts["Cycle"], y=cycle_counts["count"],
        name=name, marker_color=color, opacity=0.8
    ))
fig4.update_layout(
    title="Data points per Cycle",
    xaxis_title="Cycle", yaxis_title="Number of points",
    barmode="group", template="plotly_dark"
)
fig4.show()
```

### 4. General 2×2 subplot

```python
fig5 = make_subplots(
    rows=2, cols=2,
    subplot_titles=["Voltage vs Time", "Current vs Time", "Points per Cycle"],
    shared_xaxes=False
)
for (name, df), color in zip(dataframes.items(), colors):
    show = True  # легенда один раз
    fig5.add_trace(go.Scatter(x=df["Time [s]"], y=df["Terminal voltage [V]"],
        mode="lines", name=name, line=dict(color=color, width=1),
        legendgroup=name, showlegend=show), row=1, col=1)
    fig5.add_trace(go.Scatter(x=df["Time [s]"], y=df["Current [A]"],
        mode="lines", name=name, line=dict(color=color, width=1),
        legendgroup=name, showlegend=False), row=1, col=2)

    cycle_counts = df.groupby("Cycle").size().reset_index(name="count")
    fig5.add_trace(go.Bar(x=cycle_counts["Cycle"], y=cycle_counts["count"],
        name=name, marker_color=color, opacity=0.8,
        legendgroup=name, showlegend=False), row=2, col=1)

fig5.update_layout(
    title="Battery datasets — overview comparison",
    template="plotly_dark", height=800,
    barmode="group", hovermode="x unified"
)
fig5.show()
```
